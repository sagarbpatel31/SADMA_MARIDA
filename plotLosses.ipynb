```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
unet_xent = pd.read_csv("./runs/unet-normal/unet-normal.csv")
unet_xent_dice = pd.read_csv("./runs/unet-xent-dice/unet-xent-dice.csv")
unet_focal = pd.read_csv("./runs/unet-focal/unet-focal.csv")
attunet_xent = pd.read_csv("./runs/attunet-normal/attunet-normal.csv")
resattunet_xent = pd.read_csv("./runs/resattunet-normal/resattunet-normal.csv")
```

```python
unet_xent = unet_xent["Value"]
unet_xent_dice = unet_xent_dice["Value"]
unet_focal = unet_focal["Value"]
attunet_xent = attunet_xent["Value"]
resattunet_xent = resattunet_xent["Value"]
```

```python
unet_xent_loss = []
unet_xent_dice_loss = []
unet_focal_loss = []
attunet_xent_loss = []
resattunet_xent_loss = []
for i in range(0,1000,5):
    unet_xent_loss.append((unet_xent[i]+unet_xent[i+1]+unet_xent[i+2]+unet_xent[i+3]+unet_xent[i+4])/5)
    unet_xent_dice_loss.append((unet_xent_dice[i]+unet_xent_dice[i+1]+unet_xent_dice[i+2]+unet_xent_dice[i+3]+unet_xent_dice[i+4])/5)
    unet_focal_loss.append((unet_focal[i]+unet_focal[i+1]+unet_focal[i+2]+unet_focal[i+3]+unet_focal[i+4])/5)
    attunet_xent_loss.append((attunet_xent[i]+attunet_xent[i+1]+attunet_xent[i+2]+attunet_xent[i+3]+attunet_xent[i+4])/5)
    resattunet_xent_loss.append((resattunet_xent[i]+resattunet_xent[i+1]+resattunet_xent[i+2]+resattunet_xent[i+3]+resattunet_xent[i+4])/5)
```

```python
plt.figure(figsize=(20,10))
plt.plot(range(len(unet_xent_loss)),unet_xent_loss, label="UNet Xent Loss")
plt.plot(range(len(unet_xent_loss)),unet_xent_dice_loss, label="UNet Xent+Dice Loss")
plt.plot(range(len(unet_xent_loss)),unet_focal_loss, label="UNet Focal Loss")
plt.plot(range(len(unet_xent_loss)),attunet_xent_loss, label="AttUNet Xent Loss")
plt.plot(range(len(unet_xent_loss)),resattunet_xent_loss, label="ResAttUNet Xent Loss")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```
